```python
#####2017/12/27
from lxml import etree
import csv
import re
suggestDict = {}
warnDict = {}
errorDict = {}
Dict = {}
FILE = open('C:\Users\A501\PGhtml3.txt').read() #将数据读到文件流中
lists = FILE.split('!DOCTYPE html PUBLIC')      #按照split将txt分割成多个HTML列表
suggestSum = 0
warnSum = 0
errorSum = 0 
for list in lists:
    dom = etree.HTML(list)  #补齐缺失标记
    if dom is not None:
        suggs = dom.xpath('//ul/li[@class= "error_trp snt_hx"]') #获得suggest标记下的li信息
        for sugg in suggs:
            key = re.search(r"\[(.|\s)*?\]",etree.tostring(sugg,encoding="utf-8")).group()#正则表达式找到[key]值
            description = re.search(r"\](.|\s)*?<span",etree.tostring(sugg,encoding="utf-8")).group()
                        #正则表达式选择出描述信息内容
            if key not in suggestDict:
                suggestDict[key] = []  # 若该键值不在suggestDict列表，则添加该值
            suggestDict[key].append(description)#把描述信息分别插入suggestDict列表中
            Dict["suggest"] = suggestDict
         
        ##处理警告信息
        warns = dom.xpath('//ul/li[@class= "warn snt_hx"]')#获得warn标记下的li信息
        for warn in warns:
            key = re.search('\[[^a]*\]',etree.tostring(warn,encoding="utf-8")).group()#正则表达式找到[key]值
            description = re.search(r"\](.|\s)*?<span",etree.tostring(warn,encoding="utf-8")).group()
            #正则表达式选择出描述信息内容
            if key not in warnDict:
                warnDict[key]  = [] # 若该键值不在suggestDict列表，则添加该值
            warnDict[key].append(description)#把描述信息分别插入suggestDict列表中
        Dict["warn"] = warnDict
        
        ##处理错误信息
        errors = dom.xpath('//ul/li[@class= "error snt_hx"]')#获得error标记下的li信息
        for error in errors:
            key = re.search('\[[^a]*\]',etree.tostring(error,encoding="utf-8")).group()#正则表达式找到[key]值
            description = re.search(r"\](.|\s)*?<span",etree.tostring(error,encoding="utf-8")).group()
            if key not in errorDict:
                errorDict[key] = []   # 若该键值不在suggestDict列表，则添加该值
            errorDict[key].append(description)#把描述信息分别插入suggestDict列表中
        Dict["error"] = errorDict 
        

##将数据保存到CSV文件中
csvfile = open("PGhtml3.csv","a")
writer = csv.writer(csvfile)
for keys, values in Dict.items():
    for key, value in values.items():
        for v in value:
            writer.writerow([keys,key,v])
csvfile.close()
```
